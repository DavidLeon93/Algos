# Permutaciones en Bubblesort

Importamos las librerías que usaremos

```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
import math
```

Definimos la función de ordenamiento de Bubble

```python
def BubbleSort(a):
  comparaciones = 0
  cambios = 0
  instrucciones = 0
  i = 0
  instrucciones += 1
  while (i < len(a)):
    j = len(a) - 1
    instrucciones += 2
    comparaciones += 1
    while (i + 1 <= j):
      comparaciones += 1
      instrucciones += 1
      if (a[j] < a[j - 1]):
        temp = a[j]
        a[j] = a[j - 1]
        a[j - 1] = temp
        j -= 1
        cambios += 1
        comparaciones += 1
        instrucciones += 5
      else:
        j -= 1
        instrucciones += 1
    i += 1
    instrucciones += 1
  return (a, instrucciones, cambios, comparaciones)
```

Se crea una función para promediar una lista

```python
def promediarLista(lista):
    sum = 0
    for i in range(0, len(lista)):
        sum = sum + lista[i]

    return   sum / len( lista )
```

Para la implementación de itertools realizamos lo siguiente:

```python
listComparisonGlobal = [ ]
listExchangeGlobal = [ ]
listInstructionGlobal = [ ]

# with itertools

for i in range( 1, 12 ):
    z = i #tamano de la permutacion
    listComparison = [ ]
    listExchange = [ ]
    listInstruction = [ ]

    for x in itertools.permutations( range( z ), z ) :
        y = list ( x ) # each element x is the type 'tuple' THEN inmutable
        final = BubbleSort( y )
        #print final
        listComparison.append( final[ 1 ] )
        listExchange.append( final[ 2 ] )
        listInstruction.append( final[ 3 ] )


    listComparisonGlobal.append( promediarLista( listComparison ) )
    listExchangeGlobal.append( promediarLista( listExchange ) )
    listInstructionGlobal.append( promediarLista( listInstruction ) )
```

## Gráfica de promedio de instrucciones

```python
sizeArray = range ( 11 ) # lista del 0 al 11, porque en 11 es el número en donde la memoria colapsa
casoPromedio = listInstructionGlobal

fig = plt.figure()
axes = fig.add_subplot( 111 )
plt.bar( sizeArray, casoPromedio )
plt.ylabel('# promedio de Instrucciones' )
plt.xlabel('Numero de datos')
plt.title( " Instrucciones Promedio" )
plt.show()
```

## Gráfica de promedio de comparaciones

```python
casoPromedio = listComparisonGlobal

fig = plt.figure()
axes = fig.add_subplot( 111 )
plt.bar( sizeArray, casoPromedio )
plt.ylabel('# promedio de Comparaciones' )
plt.xlabel('Numero de datos')
plt.title( " Comparaciones Promedio" )
plt.show()
```

## Gráfica de promedio de intercambios

```python
casoPromedio = listExchangeGlobal

fig = plt.figure()
axes = fig.add_subplot( 111 )
plt.bar( sizeArray, casoPromedio )
plt.ylabel('# promedio de Intercambios' )
plt.xlabel('Numero de datos')
plt.title( " Intercambios Promedio" )
plt.show()
```
