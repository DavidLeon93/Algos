# N-ésimo término de la sucesión de Fibonacci

Se importan las librerías que usaremos

```python
import gc
import time
import matplotlib.pyplot as plt
import numpy as np
```

se define la función que determina el n-ésimo término de la sucesión de Fibonacci.

```python
def fib(n):
    x = 0
    y = 1
    aux = 0
    i = 2
    if n==0 :
        return 0
    if n==1 :
        return 1
    else: 
      while i <= n:
        aux = x + y
        x = y
        y = aux
        i += 1 
      return aux
```

se crea una lista con los tiempos para cada n de fibo(n)

```python
tiempos = []
for i in range(1000):
    gc.collect()
    tiempo_inicial = time.time()
    fib(i)
    tiempo_final = time.time()
    gc.collect()
    tiempos.append(tiempo_final - tiempo_inicial)
```

Se realiza la gráfica de los tiempos para 0<n<1000

```python
n = np.arange(0, 1000)
plt.plot(n,tiempos)
plt.xlabel("n")
plt.ylabel("Tiempo")
plt.show()
```
